```python
import pandas as pd
column_names = ["X", "Y", "result", "predicted_token"]
data = pd.read_csv('./data/output_len_1.csv', sep="#")
data.head()
```

```python
import pandas as pd

true_rows = data[data['result'] == True]
true_rows
counts = true_rows.groupby('Y').size()

# Creazione del DataFrame data_y_counts
data_y_counts = pd.DataFrame(data["Y"].value_counts()).reset_index()
data_y_counts.columns = ['Y', 'n° occorrenza']

# Creazione di una Series count
count = pd.Series(counts)

# Creazione del DataFrame count_df con conversione forzata in interi e sostituzione di NaN con 0
count_df = pd.DataFrame({'Y': count.index, 'Corrette': count.values})


# Unione dei due DataFrame
merged_df = pd.merge(data_y_counts, count_df, on='Y', how='left')
merged_df['Corrette'] = merged_df['Corrette'].fillna(0)

merged_df['% Accuratezza'] = merged_df.apply(lambda row: 0 if row['Corrette'] == 0 else round((row['Corrette'] / row['n° occorrenza']) * 100, 2), axis=1)


print(merged_df)
```
